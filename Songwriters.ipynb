```
import os
import pandas as pd

try:
  import gpt_2_simple as gpt2
except:
  !pip3 -q install gpt-2-simple
  import gpt_2_simple as gpt2
```

## Get Data

Files - GitHub : ThomasVrancken/lyrics_generation.


```
dfs = []
link = ('https://raw.githubusercontent.com/ThomasVrancken/'
        'lyrics_generation/master/songdata_{}.csv')
for i in range(4):
  dfs.append(pd.read_csv(link.format(i)))

df = pd.concat(dfs).reset_index(drop=True)
```

Create a copy of the files locally


```
if not os.path.exists('content'):
    os.makedirs('content')

pd.DataFrame({"lyrics": df['text']})\
    .to_csv(os.path.join('content', 'lyrics.csv'), index=False)
```

## Model

```
for model_name in ["124M"]:  # could be any of["124M","355M","774M"]
    gpt2.download_gpt2(model_name=model_name)
```

```
learning_rate = 0.0001
optimizer = 'adam' # adam or sgd
batch_size = 1
model_name = "124M" # match earlier choice
sess = gpt2.start_tf_sess()
```

```
gpt2.finetune(sess,
              'content/lyrics.csv',
              model_name=model_name,
              sample_every=50,
              save_every=50,
              print_every=10,
              learning_rate=learning_rate,
              batch_size=batch_size,
              restore_from='latest',
              steps=500)   # steps is max number of training steps
```

```
lst_results=gpt2.generate(
    sess,
    prefix="<|startoftext|> The sky is blue",
    nsamples=10,
    temperature=0.8,
    top_p=0.9,
    return_as_list=True,
    truncate="<|endoftext|>",
    include_prefix=True,
    )
```

```
for res in lst_results:
  print(res)
  print('\n -------//------ \n')
```
